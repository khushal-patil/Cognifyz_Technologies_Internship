```python
tabrestcuisine = pd.crosstab(rest['restaurant_name'],rest['cuisines'])
```

```python
tabrestcuisine
```

```python
def get_recommendations(cuisine_type, price_range):
    filtered_restos = rest[
        (rest['cuisines'] == cuisine_type) &
        (rest['price'] == price_range)
    ]
    filtered_names = filtered_restos['restaurant_name'].unique()
    filtered_tabrestcuisine = tabrestcuisine.loc[filtered_names]
    
    if filtered_tabrestcuisine.empty:
        return "No restaurants found matching the criteria."
    cosine_sim_matrix = cosine_similarity(filtered_tabrestcuisine.values)
    cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=filtered_tabrestcuisine.index, columns=filtered_tabrestcuisine.index)

    recommendations = []
    for restaurant_name in filtered_tabrestcuisine.index:
        sim = cosine_sim_df.loc[restaurant_name].sort_values(ascending=False)
        sim = pd.DataFrame({'restaurant_name': sim.index, 'match': sim.values})
        sim = sim[(sim['restaurant_name'] != restaurant_name) & (sim['match'] >= 0.7)]
        recommendations.append(sim)
    
    recommendations_df = pd.concat(recommendations).drop_duplicates().head(5)
    
    RestoRec = pd.merge(recommendations_df, rest[['restaurant_name', 'aggregate_rating']], how='inner', on='restaurant_name')
    FinalRestoRec = RestoRec.sort_values('aggregate_rating', ascending=False).drop_duplicates('restaurant_name', keep='first')
    
    return FinalRestoRec

cuisine_type = input("Enter the cuisine type: ")
price_range = int(input("Enter the price range eg 2000 as 2: "))

# Get recommendations
recommendations = get_recommendations(cuisine_type, price_range)
print(recommendations)
```
